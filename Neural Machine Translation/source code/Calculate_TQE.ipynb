```python
import torch
# import openpyxl
import numpy as np 
from tqe import TQE
from tqdm import tqdm
```

### Calculate TQE (Excel)

```python
GPU_NUM = 0 # 원하는 GPU 번호 입력
device = torch.device(f'cuda:{GPU_NUM}' if torch.cuda.is_available() else 'cpu')
torch.cuda.set_device(device) # change allocation of current GPU
```

```python
wb = openpyxl.load_workbook('TQE(Epoch).xlsx')
sheet = wb['model_1']
# source : 번역 원문
# target : 기계 번역 문장

for i in range(2, 1002):
    target = []
    source = []
    candidate = sheet.cell(row = i, column = 7).value.capitalize()
    reference = sheet.cell(row = i, column = 3).value
    print(i, ':', candidate)

    # Translation Quality Estimator (QE)
    # https://github.com/theRay07/Translation-Quality-Estimator
    target.append(candidate)
    source.append(reference)
    model = TQE('LaBSE')
    cos_sim_values = model.fit(source, target)
    sheet.cell(row = i, column = 8).value = cos_sim_values[0]
    
wb.save('TQE(Epoch).xlsx')
```

### Text file 버전

```python
f = open('../outputs/supermodel/hpys.txt', 'r')
outputs = f.readlines()
f.close()

f1 = open('../dataset/test/source.txt', 'r')
sources = f1.readlines()
f1.close()

t_sum, count = 0, 0

f3 = open('../outputs/supermodel/hpys_tqe.txt', 'w')
for i in range(len(outputs)):
    count += 1
    target = []
    source = []
    
    candidate = outputs[i]
    reference = sources[i]
    print(i, ':', candidate)
    
    # Translation Quality Estimator (QE)
    # https://github.com/theRay07/Translation-Quality-Estimator
    target.append(candidate)
    source.append(reference)
    model = TQE('LaBSE')
    cos_sim_values = model.fit(source, target)
    
    t_sum += cos_sim_values[0] 
    
    f3.write(str(cos_sim_values[0]))
    f3.write('\n')
    
f3.write('Average TQE: ' + str(t_sum/count)) 
f3.close()
    
```

### OOV(no word) 제거

```python
wb = openpyxl.load_workbook('TQE(5-Epoch).xlsx')
sheet = wb['model_8']

for i in range(2, 1002):
    candidate = sheet.cell(row = i, column = 8).value
    if candidate == ' ':
        sheet.cell(row = i, column = 9).value = 0


wb.save('TQE(5-Epoch).xlsx')
```

```python
wb = openpyxl.load_workbook('TQE(Epoch).xlsx')
sheet = wb['survival']
qe = []


for i in tqdm(range(2, 1002)):
    qe.append(float(sheet.cell(row = i, column = 20).value))

    
qe = np.array(qe)

sheet.cell(row = 1002, column = 20).value = qe.mean()


wb.save('TQE(Epoch).xlsx')
```
